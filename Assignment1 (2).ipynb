```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

data = pd.read_csv(r'C:\Users\SOFIA\Downloads\House Price India.csv')
```

```python
sns.histplot(data['Price'], kde=True) # Histogram
sns.kdeplot(data['Price'], fill=True) # Density plot
sns.boxplot(x=data['Price']) # Boxplot
sns.violinplot(x=data['Price']) # Violin plot
```

```python
sns.scatterplot(x=data['lot_area_renov'], y=data['Price']) # Scatter plot
sns.lineplot(x=data['lot_area_renov'], y=data['Price']) # Line plot
sns.regplot(x=data['lot_area_renov'], y=data['Price']) # Regression plot
```

```python
sns.pairplot(data, vars=['Price', 'lot area', 'number of bedrooms', 'number of bathrooms'])
```

```python
data.describe() # Descriptive statistics
```

```python
data.dropna(inplace=True) # Drop missing values
data.fillna(data.mean(), inplace=True) # Impute missing values with mean
data.fillna(data.median(), inplace=True) # Impute missing values with median
data.fillna(data.mode().iloc[0], inplace=True) # Impute missing values with mode
```
